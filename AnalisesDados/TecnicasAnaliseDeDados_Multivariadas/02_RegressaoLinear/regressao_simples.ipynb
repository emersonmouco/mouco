```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('distancia_estudantes.csv')
```

```python
print(df.head())
```

```python
plt.scatter(df['distancia_escola_minutos'], df['distancia_escola_km'])
plt.xlabel('Distância em Minutos')
plt.ylabel('Distância em Quilômetros')
plt.title('Relação entre Distância em Minutos e Quilômetros')
plt.show()
```

```python
X = df[['distancia_escola_minutos']]
y = df['distancia_escola_km']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print(f'Coeficiente de inclinação (β1): {model.coef_[0]}')
print(f'Intercepto: {model.intercept_}')
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Distância em Minutos')
plt.ylabel('Distância em Quilômetros')
plt.title('Regressão Linear: Distância em Minutos vs Quilômetros')
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print(f'Coeficiente de Determinação (R²): {r2}')
```
