```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('distancia_estudantes.csv')
```

```python
print(df.head())
```

```python
plt.scatter(df['distancia_escola_minutos'], df['distancia_escola_km'])
plt.xlabel('Dist√¢ncia em Minutos')
plt.ylabel('Dist√¢ncia em Quil√¥metros')
plt.title('Rela√ß√£o entre Dist√¢ncia em Minutos e Quil√¥metros')
plt.show()
```

```python
X = df[['distancia_escola_minutos']]
y = df['distancia_escola_km']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print(f'Coeficiente de inclina√ß√£o (Œ≤1): {model.coef_[0]}')
print(f'Intercepto: {model.intercept_}')
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Dist√¢ncia em Minutos')
plt.ylabel('Dist√¢ncia em Quil√¥metros')
plt.title('Regress√£o Linear: Dist√¢ncia em Minutos vs Quil√¥metros')
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print(f'Coeficiente de Determina√ß√£o (R¬≤): {r2}')
```
