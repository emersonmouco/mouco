```python
# verificando distribuição dos dados de indice de performance
plt.figure(figsize=(20,15))
plt.subplot(1,2,1)
sns.histplot(df['indice_performance'],kde=True)
plt.grid(axis='y')

plt.subplot(1,2,2)
plt.text(.3,.6,f"A faixa de dados de indice de performance é de:  {df['indice_performance'].max()-df['indice_performance'].min()}",fontsize=15)
plt.text(.3,.5,f"A média do índice de performance é: {df['indice_performance'].mean()}",fontsize=15)
plt.text(.3,.4,f"A mediana do índice de performance é:  {df['indice_performance'].median()}",fontsize=15)
plt.axis('off')
plt.tight_layout()
plt.show()
```
